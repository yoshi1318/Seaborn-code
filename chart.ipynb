```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start = '2020-04-01'
end = '2021-04-01'

df = data.DataReader('^N225', 'yahoo', start, end)
```

```python
df.head(10)
```

```python
df.index
```

```python
date=df.index
price=df['Adj Close']
```

```python
plt.plot(date,price)
```

```python
plt.figure(figsize=[30, 10])
plt.plot(date,price)
```

```python
span01=5
span02=25
span03=50

df['s=01'] = price.rolling(window=span01).mean()
df['s=02'] = price.rolling(window=span02).mean()
df['s=03'] = price.rolling(window=span03).mean()
```

```python
df.head(100)
```

```python
plt.figure(figsize=[30,10])
plt.plot(date,price,label='Nikkei225')
plt.plot(date,df['s=01'],label='s=01')
plt.plot(date,df['s=02'],label='s=02')
plt.plot(date,df['s=03'],label='s=03')

plt.title('N225',color='blue',backgroundcolor='white',size=40, loc='center')
plt.xlabel('date',color='black',size=30)
plt.ylabel('price',color='black',size=30)

plt.legend()
```

```python
plt.figure(figsize=[30,15])
plt.bar(date,df['Volume'],label='Volume',color='grey')

plt.legend()
```

```python
date=df.index
price=df['Close']

span01=5
span02=25
span03=50

df['s=01'] = price.rolling(window=span01).mean()
df['s=02'] = price.rolling(window=span02).mean()
df['s=03'] = price.rolling(window=span03).mean()

plt.figure(figsize=(30,15))
plt.subplot(2,1,1)

plt.plot(date,price,label='Close',color='red')
plt.plot(date,df['s=01'],label='s=01',color='yellow')
plt.plot(date,df['s=02'],label='s=02',color='green')
plt.plot(date,df['s=03'],label='s=03',color='blue')
plt.legend()

plt.subplot(2,1,2)
plt.bar(date,df['Volume'],label='Volume',color='grey')
plt.legend
```
