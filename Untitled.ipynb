```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
iris_data = sns.load_dataset('iris')
```

```python
iris_data
```

```python
sns.relplot(x='sepal_length', y='petal_length', data=iris_data)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(iris_data['sepal_length'], iris_data['petal_length'])
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
```

```python
sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, hue='species')
plt.show()
```

```python
sns.set(style='darkgrid')
sns.catplot(x='species', y='sepal_length', data=iris_data, kind='box')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips_data = sns.load_dataset('tips')
```

```python
tips_data
```

```python
sns.set(style='darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
plt.show()
```

```python
sns.set(style='darkgrid')
sns.lmplot(x="total_bill", y="tip", data=tips_data, hue="smoker")
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.histplot(data=tips_data, x='day', hue='sex', multiple='dodge')
plt.show()
```
